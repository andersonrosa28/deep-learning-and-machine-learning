```python
#Leitura dos dados
data = pd.read_excel('iris.xlsx')
```

```python
data.shape
```

```python
dados=[]
for i in range(0,150):
  dados.append([str (data.values[i,j]) for j in range (0,4)])
```

```python
print(dados)
```

```python
association_rules=apriori(dados, min_lenght=3, min_support=0.5, min_confidence=1, min_lift=1.4)
association_result=list(association_rules)
```

```python
data_result=pd.DataFrame(association_result)
data_result.head()
```
